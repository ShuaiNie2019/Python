```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['hatch.linewidth'] = 0.2

data = np.loadtxt('book1.txt', delimiter='\t')
x = data[:,0]
y = data[:,1:19]

d={}
for i in range(0,18):
        d[i]=y[:,i]
                
y1 = d[0]
for i in range(1,18):
    a = d[i]
    y1 = np.vstack([y1, a])


fig, ax = plt.subplots()
stacks = ax.stackplot(x, y1)

hatches=['-', '+', 'x', '\\', '*', 'o', 'O', '.', '/','-', '+', 'x', '\\', '*', 'o', 'O', '.', '/']
for stack, hatch in zip(stacks, hatches):
    stack.set_hatch(6*hatch)
    
ax.set_xscale('log')
ax.set(xlim=[0.001,1200])
plt.savefig('nmr.jpg',dpi=500)
plt.show()
```
