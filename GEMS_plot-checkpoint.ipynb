```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,AutoMinorLocator)


rc = {"font.family" : "Times New Roman", 
      "mathtext.fontset" : "stix"}
plt.rcParams.update(rc)
```

```python
data = pd.read_csv('Book2.txt',index_col=None, header=0,sep='\t')
```

```python
data = data.drop(columns=['AFm', 'HC','C3AH6'])
```

```python
data.head()
# data.tail()
a,b = data.shape
```

```python
data.columns[1]
```

```python

mpl.rcParams['hatch.linewidth'] = 0.2

# data = np.loadtxt('Book2.txt', delimiter='\t')

x = 10**data.iloc[:,0]
y = data.iloc[:,1:16]

d={}
for i in range(0,15):
        d[i]=y.iloc[:,i]
                
y1 = d[0]
for i in range(1,15):
    a = d[i]
    y1 = np.vstack([y1, a])
```

```python
fig = plt.figure(figsize=(6,4))
ax = plt.subplot2grid((1, 4), (0, 0), colspan=3)
stacks = ax.stackplot(x, y1,labels=data.columns[1:16])

hatches=['-', '+', 'x', '\\', '*', 'o', 'O', '.', '/','-', '+', 'x', '\\', '*', 'o', 'O', '.', '/']
for stack, hatch in zip(stacks, hatches):
    stack.set_hatch(5*hatch)
ax.legend(fontsize = 8, loc='upper right', bbox_to_anchor=(1.32, 1.02),labelspacing = 0.5,frameon = False)  
ax.yaxis.set_major_locator(MultipleLocator(30))
ax.yaxis.set_minor_locator(MultipleLocator(6))
ax.set_xscale('log')
ax.set(xlim=[0.001,1200])
ax.set(ylim=[0,150])
plt.tight_layout()
# plt.show()
plt.savefig('nmr.jpg',dpi=500)
```
